```python
#Importing the python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data from csv file and load the file into jupyter notebook
df = pd.read_csv(r"C:\Users\hp\Documents\book_sales_data.csv")
```

```python
#fetch the first five rows
df.head()
```

```python
#fetch the column names
df.columns
```

```python
#here we can fetch the numerical summary using describe function
df.describe()
```

```python
#fetch the value in Publishing Year greater than 1900 to get number min value into positive 
df = df[df["Publishing Year"] > 1900]
df.describe()
```

```python
#fetching  the null values
df.isnull().sum()
```

```python
#we can drop the null values in Book Name column 
df.dropna(subset = "Book Name",inplace=True)
```

```python
df.isnull().sum()
```

```python
#same for the language code column
df.dropna(subset="language_code",inplace=True)
```

```python
df.isnull().sum()
```

```python
#check the duplicated values 
df.duplicated().sum()
```

```python
#check the unique values
df.nunique()
```

```python
#lets create the chart for the Publishing year and Frequency columns
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequncy")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
#lets compare the genre and counts using the bar chart
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("genre")
plt.ylabel("number of books")
plt.title("Number of books in each genre")
plt.show()
```

```python
#we can find the average value by author
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False).reset_index()
```

```python
#lets create the chart for each genre and Book_ratings_count using the box plot 
sns.boxplot(x="genre", y="Book_ratings_count", data=df,color="b")
plt.xlabel("genre")
plt.ylabel("Book_ratings_count")
plt.title("Box plot of Book Ratings count for each genre")
plt.show()
```

```python
#lets create scatter plot for sale price and units sold
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("sale price")
plt.ylabel("units sold")
plt.title("Scatter plot of sale price vs units sold")
plt.show()
```

```python
language_counts = df['language_code'].value_counts()
language_counts
```

```python
#lets create pie chart for the language counts 'without' autopct
plt.pie(x=language_counts, labels=language_counts.index, startangle=90)
plt.title("Language distribution of books")
plt.show()
```

```python
#lets create pie chart for the language counts 'with' autopct
plt.pie(x=language_counts, labels=language_counts.index, startangle=90, autopct = '%1.1f%%')
plt.title("Language distribution of books")
plt.show()
```

```python
#we can find now the average  author rating by book ratings count
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False)
```

```python
#here we can find the number of language code by counts
df.groupby("language_code").size().sort_values(ascending=False)
#we have the more english language books by numbers
```

```python
#here we can find total gross sales by author
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False)
```

```python
#lets create the barplot for total gross sales 
total_gross_sales_by_author.sort_values(ascending=False).head(10).plot(kind='bar')
plt.xlabel("Author")
plt.ylabel("Total gross sales")
plt.title("Total gross sales for each author")
plt.show()
```

```python
#lets create our final chart for publishing year and units sold using line chart
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker ="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total units sold")
plt.title("Total units sold over the years")
plt.show()
```
